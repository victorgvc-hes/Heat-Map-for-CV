## Problem 1

```python
import cv2
import numpy as np
```

Declaring the path for the images:

```python
img_path = '/Users/Victor/Documents/bird.jpg'
img_path2 = '/Users/Victor/Documents/sky.jpg'
```

```python
img1 = cv2.imread(img_path)
img2 = cv2.imread(img_path2)
```

Converting to RGB and displaying the original bird image:

```python
img1=np.uint8(255 * img1)
import matplotlib.pyplot as plt
plt.imshow(img1)
plt.show()
```

Converting to RGB and displaying the original sky image:

```python
img2=np.uint8(255 * img2)
plt.imshow(img2)
plt.show()
```

Belnding the images:

```python
dst = cv2.addWeighted(img1,0.3,img2,0.3,0)
cv2.imshow('dst',dst)
cv2.waitKey(1)
cv2.destroyAllWindows()
```

Showing the images:

```python
import matplotlib.pyplot as plt
plt.imshow(dst)
plt.show()
```

We try with different weights and next we see the resulting images:

```python
dst = cv2.addWeighted(img1,0.9,img2,0.1,0)
cv2.imshow('dst',dst)
cv2.waitKey(1)
cv2.destroyAllWindows()
plt.imshow(dst)
plt.show()
```

```python
dst = cv2.addWeighted(img1,1,img2,0.2,0)
cv2.imshow('dst',dst)
cv2.waitKey(1)
cv2.destroyAllWindows()
plt.imshow(dst)
plt.show()
```

```python
dst = cv2.addWeighted(img1,0.5,img2,0.9,0)
cv2.imshow('dst',dst)
cv2.waitKey(1)
cv2.destroyAllWindows()
plt.imshow(dst)
plt.show()
```

```python
dst = cv2.addWeighted(img1,0.7,img2,0.4,0)
cv2.imshow('dst',dst)
cv2.waitKey(1)
cv2.destroyAllWindows()
plt.imshow(dst)
plt.show()
```

Subtracting the bird from the blended image:

```python
img3=cv2.subtract(img2, img1, dst)
```

```python
plt.imshow(img3)
plt.show()
```
